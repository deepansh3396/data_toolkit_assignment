```python
#Explain the following terms
#The difference in np.array, np.asarray and np.asanyarray
#The difference between Deep copy and shallow copy

# np.array()
#Purpose: Creates a new NumPy array.
#Behavior:Always creates a new array.
#Copies the data from the input, regardless of whether the input is already a NumPy array or not.
#Use Case: Use when you want to ensure you have a completely independent copy of the input data.
import numpy as np

arr1 = np.array([1, 2, 3])  # Creates a new array
```

```python
# np.asarray()
#Purpose: Converts the input to an array, but avoids copying when the input is already an array of the same type.
#Behavior:If the input is a NumPy array, it returns the same array (no copy).
#If the input is not an array, it creates a new array.
#Use Case: Use when you want to avoid unnecessary copies and the input might already be a NumPy array.
arr2 = np.asarray(arr1)  # Returns arr1 without copying if arr1 is already an array
```

```python
# np.asanyarray()
#Purpose: Similar to np.asarray(), but it allows for the return of subclasses of ndarray (like masked arrays).
#Behavior:If the input is a subclass of ndarray, it returns it without converting to a base ndarray.
#If the input is not an ndarray, it creates a new array.
#Use Case: Use when you want to maintain the type of subclasses of ndarray.
arr3 = np.asanyarray(arr1)  # Returns arr1 as is, including subclasses
```

```python
#Shallow Copy
#Definition: A shallow copy creates a new object, but it does not create copies of nested objects (like lists within lists). Instead, it copies references to those nested objects.

#Behavior:The new object is a separate object, but the elements within it are references to the same objects as those in the original.Changes to the nested objects in the shallow copy will affect the original object, as both share references to the same nested elements.
import copy

original = [[1, 2, 3], [4, 5, 6]]
shallow_copied = copy.copy(original)

shallow_copied[0][0] = 10  # Modifying the nested object
print(original)  # Output: [[10, 2, 3], [4, 5, 6]]


#Deep Copy
#Definition: A deep copy creates a new object and recursively copies all objects found within the original object, including all nested objects.

#Behavior:The new object and all nested objects are entirely independent of the original object.Changes to nested objects in the deep copy do not affect the original object.
import copy

original = [[1, 2, 3], [4, 5, 6]]
deep_copied = copy.deepcopy(original)

deep_copied[0][0] = 10  # Modifying the nested object
print(original)  # Output: [[1, 2, 3], [4, 5, 6]]
```

```python
#Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and  right y-axis for 'Humidity')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
dates = pd.date_range(start='2023-01-01', periods=30, freq='D')

# Create random temperature and humidity data
temperature = np.random.uniform(low=20, high=30, size=30)  # Random temperatures between 20 and 30
humidity = np.random.uniform(low=30, high=80, size=30  )   # Random humidity between 30% and 80%

# Create a DataFrame
data = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Set the date as the index
data.set_index('Date', inplace=True)

# Plotting
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(data.index, data['Temperature'], color='red', label='Temperature (°C)')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(data.index, data['Humidity'], color='blue', label='Humidity (%)')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Set the title
plt.title('Daily Temperature and Humidity')

# Show the plot
plt.show()
```

```python
#b) Overlay a line plot representing the normal distribution's probability density function (PDF)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, color='blue', alpha=0.7, edgecolor='black')

# Create x values for the PDF
x = np.linspace(-4, 4, 1000)
# Calculate the PDF using the normal distribution
pdf = norm.pdf(x, loc=0, scale=1)

# Overlay the PDF on the histogram
plt.plot(x, pdf, color='red', linewidth=2, label='Normal Distribution PDF')

# Set titles and labels
plt.title('Histogram of Normally Distributed Data with PDF Overlay')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```

```python
#Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Define a function to categorize quadrants
def get_quadrant(row):
    if row['x'] >= 0 and row['y'] >= 0:
        return 'Quadrant I'
    elif row['x'] < 0 and row['y'] >= 0:
        return 'Quadrant II'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Apply the function to create a new column for quadrants
data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='viridis', style='Quadrant', s=100)

# Set titles and labels
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend
plt.legend(title='Quadrants')

# Show the plot
plt.grid()
plt.show()
```

```python
#With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare output to display inline in a Jupyter Notebook (if applicable)
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2*pi
y = np.sin(x)

# Create a new figure
p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)

# Add a line glyph for the sine wave
p.line(x, y, line_width=2, color='blue')

# Add grid lines
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

```python
#Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Prepare output to display inline in a Jupyter Notebook (if applicable)
output_notebook()

# Generate random categorical data
categories = [f'Category {i}' for i in range(1, 11)]
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame for better handling
data = pd.DataFrame({'categories': categories, 'values': values})

# Generate colors based on values and add as a column to the DataFrame
data['colors'] = ['#%02x%02x%02x' % (int(v * 2.55), 0, 0) for v in values]  # Scale red based on values

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a new figure
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values', width=800, height=400)

# Add bars with color based on values, referencing the 'colors' column in the source
p.vbar(x='categories', top='values', width=0.9, source=source, color='colors') # Changed to 'colors'

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show grid lines
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

```python
#Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.

import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.rand(100)  # 100 random values

# Create a line plot
fig = go.Figure()

# Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data', line=dict(color='blue')))

# Set the title and labels
fig.update_layout(title='Simple Line Plot',
                  xaxis_title='X-axis',
                  yaxis_title='Y-axis')

# Show the plot
fig.show()
```

```python
#Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

import plotly.graph_objects as go
import numpy as np

# Generate random data for the pie chart
np.random.seed(0)  # For reproducibility
labels = [f'Category {i}' for i in range(1, 6)]  # 5 categories
values = np.random.randint(1, 100, size=5)  # Random values between 1 and 100

# Create a pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values,
                               textinfo='label+percent',  # Show labels and percentages
                               hole=0.3)])  # Optional: create a donut chart

# Set the title
fig.update_layout(title='Interactive Pie Chart')

# Show the plot
fig.show()
```
